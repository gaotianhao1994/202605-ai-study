## 1.3 第一个 LangChain 程序

现在让我们运行第一个 LangChain 程序！

```python
# ============================================
# 示例1：第一个 LangChain 程序
# ============================================

import os
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

# 加载环境变量
load_dotenv()

# 创建模型实例
# ChatOpenAI 是 LangChain 中用于调用 OpenAI Chat 模型的类
# 使用阿里云百炼的配置
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY")
)

# 调用模型并获取响应
# invoke() 方法用于调用模型，传入一个字符串作为提示词
response = llm.invoke("你好，请用一句话介绍 LangChain")

# 打印结果
print("=" * 50)
print("AI 的回答:")
print("=" * 50)
print(response.content)
print("=" * 50)
```

### 代码解释

让我们来理解这段代码：

1. **导入模块**
   - `ChatOpenAI`: LangChain 中用于调用 OpenAI Chat 模型的类
   - `load_dotenv`: 用于加载 .env 文件中的环境变量

2. **创建模型实例**
   - `llm = ChatOpenAI(...)`: 创建一个 ChatOpenAI 实例
   - 使用阿里云百炼的配置（从环境变量读取）

3. **调用模型**
   - `llm.invoke("你的问题")`: 调用模型并获取响应
   - 返回一个 AIMessage 对象

4. **获取结果**
   - `response.content`: 获取 AI 的回答内容

### 知识点：ChatOpenAI 的常用参数

**是什么？**
ChatOpenAI 支持多种参数来控制模型行为。

**为什么重要？**
不同的参数会影响输出的质量和风格。

**常用参数：**
- `model`: 模型名称（从环境变量读取）
- `openai_api_base`: API 端点（阿里云百炼）
- `openai_api_key`: API 密钥
- `temperature`: 创造性程度（0-1，越大越有创造性）
- `max_tokens`: 最大输出长度
- `timeout`: 请求超时时间

```python
# ============================================
# 示例2：使用参数创建模型实例
# ============================================

import os
from langchain_openai import ChatOpenAI

# 创建模型实例，设置参数
# 使用阿里云百炼的配置
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7,             # 创造性程度（0=保守，1=创造）
    max_tokens=500               # 最大输出长度
)

# 调用模型
response = llm.invoke("请写一个关于人工智能的短句")

print("=" * 50)
print("AI 的回答:")
print("=" * 50)
print(response.content)
print("=" * 50)
```

## 2.1 LLM vs Chat Model

### 知识点：LLM 和 Chat Model 的区别

**是什么？**
- **LLM**: 传统的语言模型，输入字符串，输出字符串
- **Chat Model**: 聊天模型，输入消息列表，输出消息

**为什么重要？**
Chat Model 更适合对话场景，可以处理多轮对话和角色扮演。

**LangChain 的选择：**
LangChain 主要使用 Chat Model，因为它更灵活、更强大。

## 2.2 不同的调用方式

LangChain 提供了多种调用方式，适用于不同的场景。

### 2.2.1 invoke() - 单次调用

最常用的调用方式，用于单次请求。

```python
# ============================================
# 示例3：使用 invoke() 进行单次调用
# ============================================

import os
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

# 单次调用
response = llm.invoke("什么是机器学习？")

print("AI 的回答:")
print(response.content)
```

### 2.2.2 batch() - 批量调用

用于一次处理多个请求，提高效率。

```python
# ============================================
# 示例4：使用 batch() 进行批量调用
# ============================================

import os
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

# 批量调用 - 一次处理多个问题
questions = [
    "什么是 Python？",
    "什么是 JavaScript？",
    "什么是 Go？"
]

responses = llm.batch(questions)

# 打印所有回答
for i, response in enumerate(responses, 1):
    print(f"\n{'='*50}")
    print(f"问题 {i}: {questions[i-1]}")
    print(f"{'='*50}")
    print(response.content)
```

### 2.2.3 stream() - 流式输出

用于实时显示输出，适合长文本生成。

```python
# ============================================
# 示例5：使用 stream() 进行流式输出
# ============================================

import os
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

print("AI 正在回答（流式输出）:")
print("=" * 50)

# 流式输出 - 实时显示每个 token
for chunk in llm.stream("请写一个关于人工智能的短故事"):
    print(chunk.content, end="", flush=True)

print("\n" + "=" * 50)
```

## 2.3 模型参数的作用

### 知识点：temperature 参数

**是什么？**
temperature 是控制 LLM 输出随机性的参数，取值范围 0-1。

**为什么重要？**
- temperature=0：输出最确定、最一致，适合事实性任务
- temperature=1：输出最有创造性、最随机，适合创意任务

**示例对比：**
- 任务："写一个产品标语"
  - temperature=0："高质量产品，值得信赖"
  - temperature=0.9："让生活更精彩，让选择更简单！"

```python
# ============================================
# 示例6：对比不同 temperature 的效果
# ============================================

import os
from langchain_openai import ChatOpenAI

# 创建两个不同 temperature 的模型
llm_conservative = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.0   # 保守
)
llm_creative = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=1.0       # 创造性
)

prompt = "写一个关于咖啡的广告语"

# 保守的回答
print("temperature=0.0（保守）:")
print("-" * 50)
response1 = llm_conservative.invoke(prompt)
print(response1.content)

print("\n" + "=" * 50 + "\n")

# 创造性的回答
print("temperature=1.0（创造性）:")
print("-" * 50)
response2 = llm_creative.invoke(prompt)
print(response2.content)
```

# 🔗 第四章：链式调用（Chains）

## 学习目标

- 理解链的概念
- 学习 LCEL 语法
- 构建简单和复杂的处理链

## 4.1 什么是链？

### 知识点：链的概念

**是什么？**
链是将多个组件串联起来形成处理流程的方式。

**为什么重要？**
- 可以实现复杂的工作流
- 代码更简洁、更易读
- 方便复用和组合

**简单比喻：**
链就像流水线，每个环节处理数据，然后传递给下一个环节。

## 4.2 LCEL - LangChain Expression Language

LCEL 是 LangChain 的表达式语言，使用 `|` 操作符连接组件。

```python
# ============================================
# 示例11：使用 LCEL 创建简单链
# ============================================

import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

# 创建组件
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)
prompt = ChatPromptTemplate.from_template("给我讲一个关于{topic}的笑话")

# 使用 | 操作符连接组件
# 链的执行顺序：prompt -> llm
chain = prompt | llm

# 调用链
response = chain.invoke({"topic": "程序员"})

print("AI 的回答:")
print(response.content)
```

## 4.3 添加输出解析器

可以使用输出解析器来格式化输出。

```python
# ============================================
# 示例12：添加输出解析器
# ============================================

import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 创建组件
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)
prompt = ChatPromptTemplate.from_template("给我讲一个关于{topic}的笑话")
output_parser = StrOutputParser()  # 将输出转换为字符串

# 创建链：prompt -> llm -> output_parser
chain = prompt | llm | output_parser

# 调用链
result = chain.invoke({"topic": "人工智能"})

print("AI 的回答:")
print(result)
```

- 学习不同的记忆类型

## 5.1 为什么需要记忆？

### 知识点：LLM 的无状态特性

**是什么？**
LLM 是无状态的，每次调用都是独立的，不会记住之前的对话。

**为什么重要？**
为了实现多轮对话，需要手动管理对话历史。

**解决方案：**
LangChain 提供了记忆组件，自动管理对话历史。

## 5.2 ConversationBufferMemory

ConversationBufferMemory 会保存完整的对话历史。

```python
# ============================================
# 示例13：使用 ConversationBufferMemory
# ============================================

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

import os
# 创建模型
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

# 创建记忆组件
memory = ConversationBufferMemory()

# 创建对话链
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True  # 显示详细过程
)

# 第一轮对话
print("第一轮对话:")
response1 = conversation.predict(input="你好，我叫小明")
print(response1)

print("\n" + "="*50 + "\n")

# 第二轮对话
print("第二轮对话:")
response2 = conversation.predict(input="你还记得我的名字吗？")
print(response2)
```

## 5.3 ConversationBufferWindowMemory

ConversationBufferWindowMemory 只保存最近 N 轮对话，避免历史过长。

```python
# ============================================
# 示例14：使用 ConversationBufferWindowMemory
# ============================================

from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferWindowMemory
from langchain.chains import ConversationChain

import os
# 创建模型
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

# 创建记忆组件 - 只保留最近 2 轮对话
memory = ConversationBufferWindowMemory(k=2)

# 创建对话链
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

# 进行多轮对话
print("第一轮对话:")
response1 = conversation.predict(input="我喜欢吃苹果")
print(response1)

print("\n第二轮对话:")
response2 = conversation.predict(input="我喜欢吃香蕉")
print(response2)

print("\n第三轮对话:")
response3 = conversation.predict(input="我喜欢吃什么水果？")
print(response3)
```

## 6.2 项目实现

让我们一步步构建这个智能学习助手。

```python
# ============================================
# 综合实战：智能学习助手
# ============================================

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

import os
# 步骤1：设计提示词模板
# 定义 AI 的角色和行为
prompt = ChatPromptTemplate.from_messages([
    ("system", """你是一个友好的学习助手，专门帮助学生学习编程和 AI 相关知识。
你的特点：
- 回答简洁明了，避免过于复杂
- 提供具体的例子和代码
- 鼓励学生动手实践
- 记住之前的对话内容，提供连贯的帮助"""),
    ("user", "{input}")
])

# 步骤2：创建模型实例
# 使用阿里云百炼的配置
llm = ChatOpenAI(
    model=os.getenv("OPENAI_MODEL_NAME"),
    openai_api_base=os.getenv("OPENAI_API_BASE"),
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.7
)

# 步骤3：添加记忆组件
memory = ConversationBufferMemory()

# 步骤4：组装成完整链
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=False
)

print("✅ 智能学习助手已创建！")
```

```python
# ============================================
# 测试智能学习助手
# ============================================

# 模拟学习对话
test_conversations = [
    "你好！我想学习 Python，应该从哪里开始？",
    "变量是什么？能举个例子吗？",
    "函数怎么定义？",
    "你刚才说的变量和函数有什么关系？"
]

print("="*60)
print("智能学习助手对话测试")
print("="*60)

for i, user_input in enumerate(test_conversations, 1):
    print(f"\n【第 {i} 轮对话】")
    print(f"学生: {user_input}")
    
    response = conversation.predict(input=user_input)
    
    print(f"\n学习助手: {response}")
    print("-"*60)
```
